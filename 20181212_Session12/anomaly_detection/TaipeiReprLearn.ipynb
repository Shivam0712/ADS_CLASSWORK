# Autoencoder for network data

Autoencoder is a non-linear dimensionality reduction tool (deep representation learning). It is an artificial neural network, or actually two neural networks being trained simultaneously:  
- first half of the layers or first network encodes the input high-dimensional data into lower-dimensional representation space, gradually reducing dimensionality from layer to layer;  
- second half transforms it back from the low-dimensional encoded representation feature space to the original dimensionality.

The architectures of the two parts are usually symmetric as could be seen on the figure below

## Data Preprocessing

### Define useful functions

**The purpose of the below is to read a temporal network (each row is a time-slice, columns are edge weights) and run autoencoder to convert it to a smaller space of representative feature timelines**

```python
# Save Pickle
def save(filename,X):
    outfile = open(filename,'wb')
    pickle.dump(X,outfile)
    outfile.close()
```

```python
# load Pickle
def load(filename):
    infile = open(filename,'rb')
    X=pickle.load(infile)
    infile.close()
    return X
```

```python
# rearrange array
def rearrange(y):
    df_temp = pd.DataFrame(y)
    lis =[]
    for i in range(y.shape[1]):
        lis.append(str(i))
    df_temp.columns = lis
    df_temp['ref']=1
    df_temp.loc[0,'ref']=0
    lis = list(df_temp.corr().sort_values(by=['ref']).index)
    lis.remove('ref')
    return np.array(df_temp[lis])
```

```python
# Plot array as image
def plotArray(y):
    fig, ax = plt.subplots(figsize=(20, 10))
    ax.imshow(y.T, cmap = 'gray')
    ax.set_aspect('auto')
    plt.show()
```

```python
# Calculate the anomaly detection metrics
def anomalyOutput(array, arrCol):
    pval=0.2 #sensitivity of anomaly detection; percentage of the most anomalous days to take
    iterN=20
    rind=np.ones(array.shape[0], dtype = bool) #index of regular (non-outlier pods/ints)
    gm=GaussianMixture(n_components=3,n_init=100,max_iter=1000,random_state=0) #clustering model
    for i in range(iterN): #iterate
        print('Iteration {}'.format(i+1))  
        clustering=gm.fit(array[rind,0:arrCol]) #fit EM clustering model excluding outliers
        l=clustering.score_samples(array[:,0:arrCol]) #estimate likelihood for each point
        Lthres=sorted(l)[int(len(l)*pval)] #anomaly threshold
        rind0 = rind
        rind= l>Lthres #non-anomalous points
        if all(rind==rind0):
            print('Convergence in {} iterations'.format(i+1))
            break
    return 100.0*sum(l[holidays]<=Lthres)/len(holidays), 100.0*sum(l[otherevents]<=Lthres)/len(otherevents), 100.0*(sum(l<Lthres)-sum(l[holidays]<=Lthres))/(len(l)-len(holidays))
```

## Baseline Autoencoder

```python
# Split Train and Test
#x_test=X[400:,0:100]; x_train=X[0:400,0:100]; #split the data into test and train


# Split Train and Test
y = np.ones(X.shape[0]) 
x_train, x_test, y_train, y_test = train_test_split( X[:,0:100], y, test_size=0.40, random_state=42)
```

```python
x_train.shape
```

```python
# Set random seeds
np.random.seed(111)
set_random_seed(21)

#design autoencoder
input_dim = x_train.shape[1]
encoding_dim = [50,15] #sizes of the layers

compression_factor = float(input_dim) / encoding_dim[-1]
print("Compression factor: %s" % compression_factor)

#define autoencoder architecture
autoencoder = Sequential()
autoencoder.add(Dense(encoding_dim[0], input_shape=(input_dim,), activation='tanh'))

for l in range(1,len(encoding_dim)):
    autoencoder.add(Dropout(0.1))
    autoencoder.add(Dense(encoding_dim[l], input_shape=(encoding_dim[l-1],), activation='tanh'))
    
for l in range(len(encoding_dim)-1,0,-1):
    autoencoder.add(Dense(encoding_dim[l-1], input_shape=(encoding_dim[l],), activation='tanh'))

autoencoder.add(Dense(input_dim, input_shape=(encoding_dim[0],), activation='sigmoid'))

autoencoder.summary()
```

```python
# keras.callbacks.EarlyStopping(monitor='val_loss',
#                               min_delta=0,
#                               patience=0,
#                               verbose=0, mode='auto')

autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
#callbacks = [EarlyStopping(monitor='val_loss', patience=5),
#             ModelCheckpoint(filepath='best_model.h5', monitor='val_loss', save_best_only=True)]
autoencoder.fit(x_train, x_train,
                epochs=400,
                batch_size=50,
                #callbacks=callbacks, # Early stopping
                shuffle=True,
                validation_data=(x_test, x_test),verbose=1)
```
